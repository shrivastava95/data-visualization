# Imports

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Loading the dataset

```python
url = "https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/2_TwoNum.csv"
data = pd.read_csv(url)
data.head()
```

# Distributions
It is a good idea to do a little preliminary analysis of numeric variables before visualizing them. This can be done by using density plots or histograms.

```python
print(data.columns)
for col in data.columns:
    plt.hist(data[col], bins=50, edgecolor='black')
    plt.title(col)
    plt.show()
```

# Scatterplot
A great way to visualize the relation between two variables in a dataset is through a scatterplot. Additionally, it helps to have line of best fit displayed on top of the data to show the trend between the variables.

```python
sns.regplot(x=data['GrLivArea'],y=data['SalePrice'], marker='.')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.title('Regression Plot showing Relation between GrLivArea and SalePrice')
plt.show()

sns.regplot(x=data['GrLivArea'][:100],y=data['SalePrice'][:100], marker='.', lowess=True)
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.title('Loess Plot showing Relation between GrLivArea and SalePrice')
plt.show()
```

It is often a good idea to show the marginal distributions in the variables when the number of data points is too high.

```python
sns.jointplot(data=data, x='GrLivArea', y='SalePrice', marker='.')
plt.show()
```

# Overplotting
When the number of data samples is too high, there can be problems in visualising the density on the pairplot. In this case, it is a good idea to try alternatives such as kde plots to help visualize the density.

```python
g = sns.kdeplot(data=data, x="GrLivArea", y="SalePrice", fill=True, levels=15)
plt.title('KDE plot across two dimensions')
plt.show()
```

```python
fig, axs = plt.subplots(ncols=2, nrows=3)
fig.set_figheight(15)
fig.set_figwidth(10)



sns.scatterplot(data=data, x="GrLivArea", y="SalePrice", s=2, ax=axs[0][0])
axs[0][0].set_title("dot size")

sns.kdeplot(data=data, x="GrLivArea", y="SalePrice", fill=False, ax=axs[0][1])
axs[0][1].set_title("Density 2d: contour")

sns.kdeplot(data=data, x="GrLivArea", y='SalePrice', fill=True,  ax=axs[1][0])
axs[1][0].set_title("Density 2d: area")

sns.kdeplot(data=data, x="GrLivArea", y="SalePrice", fill=True, ax=axs[1][1], thresh=0, levels=100, cmap="Blues")
axs[1][1].set_title("Density 2d: raster")

axs[2][0].hexbin(x=data["GrLivArea"], y=data["SalePrice"], gridsize=33, cmap="Blues")
axs[2][0].set_xlabel("GrLivArea")
axs[2][0].set_ylabel("SalePrice")
axs[2][0].set_title("Hexbin")


axs[2][1].hist2d(x=data["GrLivArea"], y=data["SalePrice"], bins=50, cmap="Blues")
axs[2][1].set_xlabel("GrLivArea")
axs[2][1].set_ylabel("SalePrice")
axs[2][1].set_title("Histogram")
plt.show()


```
