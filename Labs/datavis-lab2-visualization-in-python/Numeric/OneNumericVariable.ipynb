# Imports

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Loading the dataset

```python
url = "https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/1_OneNum.csv"
data = pd.read_csv(url)
data.head()
```

# Histogram
A histogram is useful for visualizing the density and distribution of a variable. MOst of the house prices were in the range [0, 500] so it is appropriate to zoom in onto that scale, leaving aside outliers with high prices.

Let us try splitting the data into 40 versus 100 bins, the first choice seems good as some granularity starts to emerge at 100 bins.

```python
price_lim = 500
```

```python
plt.hist(data['price'][data['price'] < price_lim], bins=40, edgecolor="black")
plt.xlabel("price")
plt.ylabel("frequency")
plt.title("bins=40")
plt.show()
plt.hist(data['price'][data['price'] < price_lim], bins=100, edgecolor="black")
plt.xlabel("price")
plt.ylabel("frequency")
plt.title("bins=100")
plt.show()
```

# Density
Density plot is similar to the histogram, actually it is sort of like a smoothed version of the histogram if it had infinite number of bins. The smoothing can be controlled with the `bandwidth` argument.

```python
bw1, bw2 = 0.1, 0.5
sns.kdeplot(data['price'][data['price'] < price_lim], bw=bw1, fill='blue')
plt.title(f"bandwidth={bw1}")
plt.show()
sns.kdeplot(data['price'][data['price'] < price_lim], bw=bw2, fill='blue')
plt.title(f"bandwidth={bw2}")
plt.show()
```
